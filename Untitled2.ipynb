## Objective

The goal of this project is to analyze user behavior in an e-commerce platform and identify the reasons behind cart abandonment.

This includes analyzing user activity across different stages of the purchase funnel (view, add to cart, checkout, purchase) and deriving insights to improve conversion rates.

```python
import pandas as pd

df = pd.read_csv("ecommerce_events_large.csv")

df.head()
```

```python
df['event_type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['event_type'].value_counts().plot(kind='bar')
plt.title("Funnel Analysis")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.show()
```

### Insight:

There is a significant drop in users from the 'add_to_cart' stage to the 'purchase' stage, indicating a high cart abandonment rate. This suggests potential issues in checkout flow, pricing, or user experience.

```python
cart_users = df[df['event_type'] == 'add_to_cart']['user_id'].nunique()
buyers = df[df['event_type'] == 'purchase']['user_id'].nunique()

abandonment_rate = (cart_users - buyers) / cart_users * 100

print("Cart Users:", cart_users)
print("Buyers:", buyers)
print(f"Abandonment Rate: {abandonment_rate:.2f}%")
```

```python
device_users = df.groupby('device')['user_id'].nunique()

print(device_users)
```

```python
device_users.plot(kind='bar')
import matplotlib.pyplot as plt

plt.title("Users by Device")
plt.xlabel("Device")
plt.ylabel("Unique Users")
plt.show()
```

### Insight:

Mobile users dominate the platform usage. However, further analysis is required to determine if mobile users are converting at a lower rate compared to desktop users, indicating potential optimization areas.

```python
category_users = df.groupby('category')['user_id'].nunique()

print(category_users)
```

```python
category_users.plot(kind='bar')
import matplotlib.pyplot as plt

plt.title("Users by Category")
plt.xlabel("Category")
plt.ylabel("Unique Users")
plt.show()
```

### Insight:

User engagement is fairly evenly distributed across all product categories, indicating diverse customer interests and no single category dependency.

## Conclusion

The analysis reveals a significant cart abandonment rate of approximately 55%, with a major drop occurring between the add-to-cart and purchase stages.

Mobile users contribute the majority of traffic, suggesting the need for mobile-optimized checkout processes.

The insights indicate potential opportunities to improve conversion rates by optimizing the checkout experience, pricing strategies, and user interface.

This analysis can help businesses take data-driven decisions to reduce cart abandonment and increase revenue.
